```python
import requests

url = "https://api.binance.com/api/v3/ticker/price?symbol=BTCUSDT"
response = requests.get(url)
records = response.json()

def gen(n=1):
    for i in range(n):
        data = {
            "symbol": records["symbol"],
            "price": float(records["price"])
        }
        print(data)

gen()
```
